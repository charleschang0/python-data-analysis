可以透過DataFrame的Describe方法，對數值統計信息快速了解

這些交易數據均不是有效成交數據，不貢獻銷售，不再後續分析範圍內，因此我們將在數據清理步驟，將Quantity為負數的觀察值刪除
接下來篩選出UnitPrice為負數的觀察值

```python
original_data[original_data["UnitPrice"] < 0]
```

從輸出結果來看，UnitPrice為負數的觀察值都是壞帳，不屬於實際商品交易數據，因此在數據清理步驟中將其刪除。

## 清理數據

根據前面評估部分得到的結論，數據清理包括：
- 把InvoiceData 的數據類型轉換成日期時間
- 把CustomerID的數據轉換成字符串
- 把Description缺失的觀察值刪除
- 把Country值USA轉換成United States/ UK變成United Kingdom
- 把Quantity為負數的觀察值刪除
- 把UnitPrice為負數的觀察值刪除

為了區分清理的數據＆原始數據，要創建新的變量cleaned_data

```python
cleaned_data= original_data.copy()
cleaned_data.head()
```

把InvoiceDate轉換成日期時間類型：

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

把CustomerID的數據類型轉換成字符串

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

把CustomerID結尾的",0"刪除：

```python
cleaned_data["CustomerID"].str.slice(0,-2)
```

把Description缺失的觀察值刪除，並查看刪除後該列空缺值個數和 (布爾值可以被視作0, 1，所以可以用sum來查看)

```python
cleaned_data.dropna(subset=["Description"], inplace= True)
```

```python
cleaned_data["Description"].isnull().sum()
```

把USA轉換成United States

```python
cleaned_data["Country"]= cleaned_data["Country"].replace({"USA": "United States"})
cleaned_data["Country"]= cleaned_data["Country"].replace({"UK": "United Kingdom", "U.K.": "United Kingdom"})
```

```python
len(cleaned_data[cleaned_data["Country"] == "U.K."])
```

把Quantity為負數的觀察值刪除，並檢查替換後Quantity為負數的個數（方法：篩選出正數，並重新賦值給cleaned_data）

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >=0]
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >=0]
```

```python
len(cleaned_data[cleaned_data["Quantity"]<0])
```

## 保存清理後的數據

```python
cleaned_data.to_csv("e_commerce_cleaned.csv", index= "False")
```
